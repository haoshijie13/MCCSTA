```r
library(ggplot2)
library(sf)
Sys.setenv(PROJ_LIB = "/mnt/SSD4Ta/home/huangzhi/anaconda3/envs/work/share/proj")
```

```r
Archr_col <- c('#D51F26','#272E6A','#208A42','#89288F','#F47D2B','#FEE500','#8A9FD1',
               '#C06CAB','#E6C2DC','#90D5E4','#89C75F','#F37B7D','#9983BD','#D24B27',
               '#3BBCA8','#6E4B9E','#0C727C','#7E1416','#D8A767')
```

```r
flatmap_mar <- read_sf('data/MRI_shp_file/Marmoset.shp')
flatmap_mac <- read_sf('data/MRI_shp_file/Macaque.shp')
flatmap_hum <- read_sf('data/MRI_shp_file/Human.shp')
flatmap_mou <- read_sf('data/MRI_shp_file/Mouse.shp')
```

```r
options(repr.plot.width=20,repr.plot.height=10)
#Plot marmoset
p1 <- ggplot(flatmap_mar) +
geom_sf(aes(fill=flatmap_mar$VL),color='black',lwd=0.1,show.legend = T)+
scale_fill_manual(values = Archr_col,breaks = sort(unique(flatmap_mar$VL)))+
geom_sf_text(aes(label = region),size=6)+
theme_void()+
theme(text=element_text(size=16))
p1
```
